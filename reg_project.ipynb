```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("data/homeprices.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color="red",marker="+")
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#y=m*x+b
135.78767123*3300+180616.43835616432
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color="red",marker="+")
plt.plot(df.area,reg.predict(df[['area']]),color="blue",marker="+")
```

```python
d=pd.read_csv("data/areas.csv")
d.head()
```

```python
p=reg.predict(d)
p
```

```python
d['prices']=p
```

```python
d
```

```python
d.to_csv("predictions.csv",index=False)
```
